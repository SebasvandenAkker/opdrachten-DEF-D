```python
# Zet hier jouw data in een voor jou handig formaat
import numpy as np
telefoon = np.array([1318,943,763,672,527,450,381])
sensor = np.array([289,250,207,183,169,149,130])
```

```python
# Plot hier de meting en fit een lijn door de metingen.
# Maak gebruik van wat je geleerd hebt bij het practicum.
%matplotlib inline
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
plt.scatter(telefoon, sensor)
plt.errorbar(telefoon, sensor, 10, 40, "*")
plt.xlabel("Telefoon metingen lux")
plt.ylabel("Sensor metingen #")

def f(telefoon, a, b):
    return a*telefoon + b

val, cov = curve_fit(f, telefoon, sensor)
x_fit = np.linspace(min(telefoon), max(telefoon), 100)
y_fit = f(x_fit, val[0], val[1])
plt.plot(x_fit,y_fit)

print(val)
```
